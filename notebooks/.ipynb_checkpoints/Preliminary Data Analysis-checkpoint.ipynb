# Timezone Helpers

```python
def convert_to_sgt(s):
    """
    s: datetime string
    """
    tz = pytz.timezone('Asia/Singapore')
    dt = datetime.strptime(s, '%Y-%m-%d %H:%M:%S')
    dt.replace(tzinfo=pytz.UTC)
    return dt.replace(tzinfo=tz)
```

What would be a better plot to see if we have the data that we need to get? Perhaps the same heat map for a **single** time slice would make more sense. Let's try to segment our data by the time that we scraped it for.

```python
len(data.created_at.unique())
```

### Bike distribution according to our time slice

```python
time_slices = sorted(data.created_at.unique())

def get_data_at_slice(i, data):
    return data[data['created_at'] == time_slices[i]]

def plot_at_slice(i, data):
    time_slice = get_data_at_slice(i, data)
    title = "Distribution of bikes at time {}".format(time_slices[i])
    g = sns.jointplot(x='longitude', y='latitude', data=time_slice, kind='kde', stat_func=None)
    plt.show()
```

```python
plot_at_slice(0, data)
```

```python
plot_at_slice(1, data)
```

```python
plot_at_slice(2, data)
```

```python
plot_at_slice(20, data)
```

```python
plot_at_slice(50, data)
```

We can observe the distribution moving, maybe if we animate this it will be a bit more clearer how the bikes are moving over time.

```python
fig, ax = plt.subplots(figsize=(6, 6))

ax.set_xlim((data.longitude.min(), data.longitude.max()))
ax.set_ylim((data.latitude.min(), data.latitude.max()))

line, = ax.plot([], [], lw=2)

def init():
    line.set_data([], [])
    return (line,)

def animate(i):
    slice_data = get_data_at_slice(i, data)
    sns.kdeplot(slice_data.longitude, slice_data.latitude, shade=True)
    return (line, )
    
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=117, interval=100, blit=True, repeat=True)
rc('animation', html='html5')
HTML(anim.to_html5_video())
```

Not very particularly interesting so far. But we can see bikes moving from one point to another. Perhaps with even more data, we can observe a better pattern?
